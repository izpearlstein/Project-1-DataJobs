```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Load the Indeed file

file = 'indeed_data.csv'

df = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

df.head()

df [df['position'] == 'Software Engineer, Machine Learning']
```

```python
#Split location to get city, state and zipcode

df [['City', 'State', 'Zip Code', 'Trash', 'Trash_1']] = df['location'].str.split(' ', expand=True)

df.head()
```

```python
# remove comma from City

df['City'] = df['City'].str.replace(',', '')

df.head()
```

```python
# delet unneccesary columns

del df['Trash']
del df['Trash_1']
del df['location']
del df['reviews']
del df['description']

df.head(20)
```

```python
# Make city upper case

df['City'] = df['City'].str.upper()

df.head(10)
```

```python
# Display an overview of the Position column

df['position'].value_counts()
```

```python
# Clean up position category. Replace for data scientist
df['position'] = df['position'].replace(
    {'Senior Data Scientist': 'Data Scientist', 'Scientist': 'Data Scientist', 'Research Scientist': 'Data Scientist', 
     'Sr. Data Scientist': 'Data Scientist', 'Lead Data Scientist': 'Data Scientist', 'Principal Data Scientist': 'Data Scientist', 
     'Bioinformatics Scientist': 'Data Scientist', 'Staff Data Scientist': 'Data Scientist', 'Associate Scientist': 'Data Scientist',
     'Senior Scientist': 'Data Scientist', 'Machine Learning Scientist': 'Data Scientist', 'Director, Data Science': 'Data Scientist', 
    'Senior Data Analyst': 'Data Scientist'})

# clean up postion category. Replace for Data Engineer

df['position'] = df['position'].replace(
    {'Machine Learning Engineer': 'Data Engineer', 'Senior Software Engineer': 'Data Engineer', 'Senior Data Engineer': 'Data Engineer',
    'Lead Data Engineer': 'Data Engineer', 'Sr. Data Engineer': 'Data Engineer'})

# Clean up postion category. replace for data analyst

df['position'] = df['position'].replace(
    {'Quantitative Analyst': 'Data Analyst', 'Market Research Analyst': 'Data Analyst', 'Senior Research Analyst': 'Data Analyst',
    'Research Analyst': 'Data Analyst', 'Research Data Analyst': 'Data Analyst'})

# Clean up position category. replace for software engineer

df['position'] = df['position'].replace(
    {'Software Engineer, Machine Learning': 'Software Engineer', 'Software Development Engineer': 'Software Engineer'
    })

# Clean up position category. replace for business analyst

df['position'] = df['position'].replace(
    {'Product Manager': 'Business Analyst'
        
    })

df['position'].value_counts()
```
