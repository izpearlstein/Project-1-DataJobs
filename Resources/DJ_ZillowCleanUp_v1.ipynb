```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# name of file

ZillowFile = 'Zillow_Data.xlsx'

ZipFile = 'Zipcode_v1.xlsx'
```

```python
#Load zillow file

df = pd.read_excel('Zillow_Data.xlsx', sheetname='Housing')

zip_df = pd.read_excel('Zipcode_v1.xlsx', sheetname = 'Sheet1', dtype={'Zip Code': 'str'}) 
```

```python
#view zip head

zip_df.head()

```

```python
# view zillow head

df.head()
```

```python
#Create average of Median Housing Price price YTD

df['Average of Median Housing Price YTD'] = (df['2019-01']+df['2019-02']+df['2019-03']+df['2019-04']+df['2019-05']+df['2019-06'])/6

df.head()
```

```python
#Split Region Name by ','

df [['RegionName', 'State']] = df['RegionName'].str.split(',', expand=True)

df.head()
```

```python
# Clean up columns to have only those that are necessary

del df['2019-01']
del df['2019-02']
del df['2019-03']
del df['2019-04']
del df['2019-05']
del df['2019-06']
del df['RegionID']
del df ['SizeRank']

# change column names

df.columns = ['City', 'Median Housing Price YTD', 'State']

df.head()
```

```python
#Split City by '-'

df [['City', 'Trash','Trash_1']] = df['City'].str.split('-', expand=True)

df.head()
```

```python
# get rid of the splits

del df['Trash']
del df['Trash_1']

df.head()
```

```python
#Remove white space from state

df['State'] = df['State'].str.replace(' ', '')

zip_df['State'] = zip_df['State'].str.replace(' ','')
```

```python
# Make city upper case

df['City'] = df['City'].str.upper()

df.head()
```

```python
# the big merge

#merge_df = pd.merge(df, zip_df,  how='left', left_on=['City','State'], right_on = ['City','State'])

merge_df = pd.merge(df, zip_df, how='left', on=['City', 'State'])

merge_df.head(20)

```

```python
# drop nan's

merge_df.dropna()

merge_df.head(20)
```

```python
# Save cleaned data to csv

df.to_csv("Cleaned_Zillow_prior2ZipJoin.csv", index=False)
```
