# Requests Text Analysis 

```python
import requests
from collections import Counter
```

```python
response = requests.get ("https://jobs.github.com/positions.json?description=data&location=us")
```

```python
response = response.json()
```

```python
# The following is a variable which is a list of lists with all the "description" of the response dictionary:

descriptions = [n["description"] for n in response]
```

```python
import re
```

```python
# The following function is going to take in a sentence and it is going to split the sentence by words. 
# It will also stip punctuation and whitespace.
# It will return the sentence with each word separated. 

def cleanSentence(sentence):
    sentence = sentence.split (" ")
    sentence = [word.strip(string.punctuation + string.whitespace) for word in sentence]
    sentence = [word for word in sentence if len(word) > 1 or (word.lower()== "a" or word.lower()== "i")]
    return sentence 


# This function is going to take in the entire text of each description and it is going to remove html tags
# It will also separate the text every period.
# At the end, it is going to send every sentence to the function "cleanSentence", which will return every sentence
# separated by words. 

def CleanInput (text):
    text = re.sub("(</p>+|<ul>+|<li>+|< >+|\n+|</li>+|</ul>+|<p>+|/+|-+ )", " ", text)
    text = bytes ( text, "UTF-8")
    text = text.decode ("ascii", "ignore")
    sentences = text.split(".")
    return [cleanSentence(sentence) for sentence in sentences]


# This function is going to create the Ngrams

"""getNgramsFromSentence: """

def CreateNgramsFromSentence (sentences, n): 
    output = []
    for s in sentences: 
        for i in range(len(s)-n+1):
                output.append (s[i:i+n])
    return output

def getNgrams (content, n):
    content = CleanInput (content)
    ngrams = Counter()
    newNgrams = [" ".join(ngram) for ngram in CreateNgramsFromSentence (content,n)]
    
    ngrams.update (newNgrams)
    #ngrams.extend (CreateNgramsFromSentence(content, n))
    return (ngrams)
```

```python
ngrams = [getNgrams (x, 2) for x in descriptions]
```
